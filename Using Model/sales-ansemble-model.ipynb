```python
paths = [x for x in glob.glob('../input/dataset/*.csv') if 'dataset' in x]
paths
```

```python
weights = [1.5, 1.6, 2.0, 0.3,0]
```

```python
dfs = [pd.read_csv(x) for x in paths]
dfs = [x.sort_values(by='id') for x in dfs]
```

```python
for df in dfs:
    df['sales'] = np.clip(df['sales'],0, 1000)
#for df in dfs:
#    df['sales'] = np.clip(df['sales'], 0, 1)
```

```python
submission = pd.read_csv('../input/store-sales-time-series-forecasting/sample_submission.csv')
submission.head()
```

```python
for df, weight in zip(dfs, weights):
    submission['sales'] += (df['sales'] * weight)
```

```python
submission['sales'] /= np.sum(weights)
```

```python
submission.to_csv('ensemble.csv', index=None)
```
